```python
all_pages_data =[]

def main():
    number_of_pages =  int(driver.find_elements(By.XPATH, "//*[contains(@class,'d6cvqb BBwThe')] /preceding-sibling::td[1]/a[1]")[0].text.replace('"', ""))
    
    for j in range(number_of_pages - 1):
        
        all_pages_data.extend(get_all_data())
        driver.find_elements(By.XPATH, "//*[contains(@id,'pnnext')]")[0].click()
        
        time.sleep(10)
            
    
if __name__ == "__main__":
    main()
```
